```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objects as go
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv(r'C:\Users\xinya\PhD\UIUC Courses\IS 537\Project\archive\movies.csv')
df
```

```python
# Drop nan values
df.dropna(inplace=True)
df.head()
```

```python
# get rid of \n
df['GENRE'] = df['GENRE'].str.replace('\n', '')
df['ONE-LINE'] = df['ONE-LINE'].str.replace('\n', '')
df['STARS'] = df['STARS'].str.replace('\n', '')
df.head()
```

```python
# Separate multiple values in a cell
df['GENRE'] = df['GENRE'].str.split(', ')
df = df.explode('GENRE')
# get rid of blank space in GENRE column
df['GENRE'] = df['GENRE'].str.strip()
df.head()
```

```python
# Turn str into int
df['VOTES'] = df['VOTES'].str.replace(',', '').astype(int)
df.head()
```

```python
# Turn Gross column into float
df['Gross'] = df['Gross'].str.replace('$', '').str.replace('M', '',regex=False).astype(float)
df
```

```python
# Separate director and stars
df[['Director', 'Stars']] = df['STARS'].str.split('|', expand=True)
df['Director'] = df['Director'].str.split(':', expand=True)[1]
df['Stars'] = df['Stars'].str.split(':', expand=True)[1]
df['Stars'] = df['Stars'].str.split(', ')
df = df.explode('Stars')
df.drop('STARS', axis=1, inplace=True)
df.head()
```
